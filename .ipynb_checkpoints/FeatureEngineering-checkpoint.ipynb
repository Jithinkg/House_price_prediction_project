```python
#replacing missing value with a new label
def replace_cat_feature(dataset,ftr_nan):
    data=dataset.copy()
    data[ftr_nan]=data[ftr_nan].fillna('Missing')
    return data
dataset=replace_cat_feature(dataset,ftr_nan)
dataset[ftr_nan].isnull().sum()
    
    
```

```python
dataset.head()
```

```python
#handling numerical variables
numer_nan=[ftr for ftr in dataset.columns if dataset[ftr].isnull().sum()>1 and dataset[ftr].dtypes!='O']
for ftr in numer_nan:
    print(ftr,":        " ,np.round(dataset[ftr].isnull().mean(),4),  '% missing values')
```

```python
for ftr in numer_nan:
    #we replcace by using median since there are outliers
    median_value=dataset[ftr].median()
    
    #create new feature to capture nan values
    dataset[ftr+'nan']=np.where(dataset[ftr].isnull(),1,0)
    dataset[ftr].fillna(median_value,inplace=True)
dataset[numer_nan].isnull().sum()
```

```python
dataset.head()
```

```python
#handling temporal variables
for ftr in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[ftr]=dataset['YrSold']-dataset[ftr]
#dataset[ftr].dtypes
dataset.head()
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
#now we handle those numerical varibale(cont. var) having no 0 values.
import numpy as np
num_ftr=['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']

for ftr in num_ftr:
    dataset[ftr]=np.log(dataset[ftr])
dataset.head()
```

```python
cat_ftr=[ftr for ftr in dataset.columns if dataset[ftr].dtype=='O']
cat_ftr
```

```python
for feature in cat_ftr:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(100)
```

##feature scaling

```python
for feature in cat_ftr:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
    
```
